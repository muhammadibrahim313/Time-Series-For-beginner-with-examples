# Introduction to Time Series 

**What is Time Series Data?**

Time series data is a sequence of data points collected or recorded at successive points in time. Each data point is associated with a specific timestamp. Examples include daily stock prices, monthly sales figures, or hourly temperature readings.

**Why is it Important?**
Time series data is crucial because it helps us understand how a variable changes over time, identify patterns or trends, and make future predictions based on past behaviors. This is essential in fields such as finance, weather forecasting, and economics.

**Examples:**
- **Stock Prices:** Daily closing prices of a company's stock show how the value of the stock changes over time.
- **Daily Temperature Readings:** Daily high and low temperatures help track weather patterns and climate changes.
- **Monthly Sales Figures:** Monthly sales data can show seasonal trends and help businesses forecast future sales.

### Real-world Examples

**Finance:** 
- **Stock Market Analysis:** Investors use historical stock prices to predict future price movements and make investment decisions.

**Meteorology:** 
- **Weather Forecasting:** Meteorologists use past weather data to predict future weather conditions, helping prepare for storms, droughts, and other weather events.

**Economics:** 
- **GDP Growth Rates:** Economists analyze quarterly GDP data to understand economic trends, set policies, and forecast future economic performance.

## Basic Concepts

### Components of Time Series

**Trend:**
- The overall direction of the data over a long period. It shows whether the data is increasing, decreasing, or staying constant over time.
- **Example:** An increasing trend in annual average temperatures over several decades indicates global warming.

**Seasonality:**
- Regular patterns that repeat at fixed intervals, such as daily, monthly, or yearly.
- **Example:** Higher ice cream sales in summer and lower sales in winter due to seasonal preferences.

**Cyclicity:**
- Long-term cycles in the data that are not of fixed length and often influenced by economic or business cycles.
- **Example:** Economic expansions and recessions.

**Irregularity:**
- Random variations or noise that do not follow a pattern and cannot be attributed to trend, seasonality, or cycles.
- **Example:** Sudden stock price drops due to unexpected news.

### Types of Time Series

**Stationary vs. Non-stationary:**
- **Stationary:** A time series whose statistical properties, like mean and variance, remain constant over time. 
  - **Example:** White noise, where each data point is random but the overall distribution remains the same.
- **Non-stationary:** A time series whose statistical properties change over time.
  - **Example:** Stock prices that show a long-term increasing trend.

### Visualization Techniques

**Plotting Time Series Data:**
- **Line Plots:** A simple and effective way to visualize time series data, where time is on the x-axis and the variable of interest is on the y-axis.

### **Example** 

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('/kaggle/input/tesla-stock-data-updated-till-28jun2021/TSLA.csv', parse_dates=['Date'], index_col='Date')
data['Close'].plot()
plt.title('Stock Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.show()
```

**Highlighting Trends and Seasonal Components:**
- Use moving averages or decomposition techniques to visualize and highlight trends and seasonality in the data.

## Time Series Decomposition

### Additive and Multiplicative Models

**Additive Model:**
- Assumes that the components of the time series (trend, seasonality, and irregularity) are added together.
- **Formula:** \( Y(t) = T(t) + S(t) + R(t) \)
- Use when seasonal variations are roughly constant over time.
- **Example:** Monthly electricity consumption that shows a steady increase over time with a constant seasonal pattern.

**Multiplicative Model:**
- Assumes that the components of the time series multiply together.
- **Formula:** \( Y(t) = T(t) \times S(t) \times R(t) \)
- Use when seasonal variations change proportionally with the level of the series.
- **Example:** Retail sales with growing seasonal peaks as the overall sales volume increases.

### Decomposition Methods

**Moving Averages:**
- A technique to smooth out short-term fluctuations and highlight longer-term trends or cycles.
- **Example:**

```python
data['Moving_Average'] = data['Close'].rolling(window=12).mean()
data[['Close', 'Moving_Average']].plot()
plt.title('Moving Average')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

**STL Decomposition:**
- Seasonal-Trend decomposition using Loess (STL) is a method to break down the time series into seasonal, trend, and residual components.


## Stationarity and Differencing

### Stationarity

**What Makes a Time Series Stationary?**
- A stationary time series has constant statistical properties over time, such as mean, variance, and autocorrelation.
- Stationarity is crucial for many time series models because it ensures that the model parameters do not change over time, leading to more reliable predictions.

**Why is it Important for Analysis?**
- Many statistical models assume stationarity to provide accurate predictions and reliable parameter estimates. If the series is not stationary, these models may produce misleading results.

### Testing for Stationarity

**Augmented Dickey-Fuller (ADF) Test:**
- A statistical test used to check for the presence of a unit root in the time series, indicating non-stationarity.
- **Example:**

```python
from statsmodels.tsa.stattools import adfuller

result = adfuller(data['Close'])
print('ADF Statistic:', result[0])
print('p-value:', result[1])
```

**KPSS Test:**
- Kwiatkowski-Phillips-Schmidt-Shin (KPSS) test is used to test for stationarity around a deterministic trend.
- **Example:**

```python
from statsmodels.tsa.stattools import kpss
import warnings
warnings.filterwarnings('ignore')

result = kpss(data['Close'])
print('KPSS Statistic:', result[0])
print('p-value:', result[1])
```

### Differencing

**How Differencing Can Make a Series Stationary:**
- Differencing involves subtracting the previous observation from the current observation to remove trends or seasonality, making the time series stationary.
- **Step-by-Step Example of Differencing:**

```python
data['Differenced'] = data['Close'].diff().dropna()
data['Differenced'].plot()
plt.title('Differenced Series')
plt.xlabel('Date')
plt.ylabel('Differenced Price')
plt.show()
```

## Smoothing Techniques

### Moving Averages

**Simple Moving Average (SMA):**
- Averages data over a specified number of periods to smooth out short-term fluctuations.
- **Example Calculation and Plot:**

```python
data['SMA'] = data['Close'].rolling(window=12).mean()
data[['Close', 'SMA']].plot()
plt.title('Simple Moving Average')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

**Weighted Moving Average (WMA):**
- Gives more weight to recent data points, making the average more responsive to changes.
- **Example Calculation and Plot:**

```python
import numpy as np

weights = np.arange(1, 13)
data['WMA'] = data['Close'].rolling(window=12).apply(lambda x: np.dot(x, weights)/weights.sum(), raw=True)
data[['Close', 'WMA']].plot()
plt.title('Weighted Moving Average')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

**Exponential Moving Average (EMA):**
- Gives more weight to recent observations using an exponentially decreasing factor, making it even more responsive to changes than WMA.
- **Example Calculation and Plot:**

```python
data['EMA'] = data['Close'].ewm(span=12, adjust=False).mean()
data[['Close', 'EMA']].plot()
plt.title('Exponential Moving Average')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

### Exponential Smoothing Methods

**Simple Exponential Smoothing (SES):**
- Suitable for data with no trend or seasonality. It assigns exponentially decreasing weights to past observations.
- **Example:**

```python
from statsmodels.tsa.holtwinters import SimpleExpSmoothing

model = SimpleExpSmoothing(data['Close']).fit()
data['SES'] = model.fittedvalues
data[['Close', 'SES']].plot()
plt.title('Simple Exponential Smoothing')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

**Holtâ€™s Linear Trend Model:**
- Suitable for data with a trend. It extends SES by adding a trend component.
- **Example:**

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing

model = ExponentialSmoothing(data['Close'], trend='add').fit()
data['Holt'] = model.fittedvalues
data[['Close', 'Holt']].plot()
plt.title('Holtâ€™s Linear Trend Model')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

**Holt-Winters Seasonal Model:**
- Suitable for data with both trend and seasonality. It extends Holtâ€™s model by adding a seasonal component.
- **Example:**

```python
model = ExponentialSmoothing(data['Close'], trend='add', seasonal='add', seasonal_periods=12).fit()
data['Holt_Winters'] = model.fittedvalues
data[['Close', 'Holt_Winters']].plot()
plt.title('Holt-Winters Seasonal Model')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

## Time Series Models

### Autoregressive (AR) Models

**Explanation:**
- Models the time series using its own past values.
- **Example:**

```python
from statsmodels.tsa.ar_model import AutoReg

model = AutoReg(data['Close'], lags=1).fit()
data['AR'] = model.fittedvalues
data[['Close', 'AR']].plot()
plt.title('Autoregressive Model')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

### Moving Average (MA) Models

**Explanation:**
- Models the time series using past forecast errors.
- **Example:**

```python
from statsmodels.tsa.arima.model import ARIMA

model = ARIMA(data['Close'], order=(0, 0, 1)).fit()
data['MA'] = model.fittedvalues
data[['Close', 'MA']].plot()
plt.title('Moving Average Model')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

### Autoregressive Moving Average (ARMA) Models

**Explanation:**
- Combines AR and MA models.
- **Example:**

```python
model = ARIMA(data['Close'], order=(1, 0, 1)).fit()
data['ARMA'] = model.fittedvalues
data[['Close', 'ARMA']].plot()
plt.title('Autoregressive Moving Average Model')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

### Autoregressive Integrated Moving Average (ARIMA) Models

**Explanation:**
- Extends ARMA to include differencing for stationarity.
- **Step-by-Step Model Building and Example:**

```python
model = ARIMA(data['Close'], order=(1, 1, 1)).fit()
data['ARIMA'] = model.fittedvalues
data[['Close', 'ARIMA']].plot()
plt.title('Autoregressive Integrated Moving Average Model')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

### Seasonal ARIMA (SARIMA) Models

**Explanation:**
- Extends ARIMA to handle seasonality.
- **Step-by-Step Model Building and Example:**

```python
model = ARIMA(data['Close'], order=(1, 1, 1), seasonal_order=(1, 1, 1, 12)).fit()
data['SARIMA'] = model.fittedvalues
data[['Close', 'SARIMA']].plot()
plt.title('Seasonal ARIMA Model')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

### State-space Models

**Kalman Filter:**
- Uses a series of measurements over time to estimate unknown variables.
- **Example:**

```python
from pykalman import KalmanFilter

kf = KalmanFilter(initial_state_mean=0, n_dim_obs=1)
state_means, _ = kf.filter(data['Close'])
data['Kalman_Filter'] = state_means
data[['Close', 'Kalman_Filter']].plot()
plt.title('Kalman Filter')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

## Model Evaluation and Selection

### Criteria for Model Selection

**AIC, BIC:**
- **AIC (Akaike Information Criterion):** Measures the goodness of fit and complexity of the model. Lower AIC values indicate better models.
- **BIC (Bayesian Information Criterion):** Similar to AIC but includes a penalty term for the number of parameters. Lower BIC values indicate better models.
- **Example:**

```python
print('AIC:', model.aic)
print('BIC:', model.bic)
```

### Cross-validation Techniques

**Explanation:**
- Split data into training and validation sets multiple times to evaluate the modelâ€™s performance.
- **Example of k-fold Cross-validation:**

```python
from sklearn.model_selection import TimeSeriesSplit

tscv = TimeSeriesSplit(n_splits=5)
for train_index, test_index in tscv.split(data):
    print("TRAIN:", train_index, "TEST:", test_index)
```

### Model Diagnostics

**Residual Analysis:**
- Check if residuals are white noise, meaning they should have a mean of zero and no autocorrelation.
- **Example:**

```python
residuals = model.resid
plt.plot(residuals)
plt.title('Residual Analysis')
plt.xlabel('Date')
plt.ylabel('Residuals')
plt.show()
```

### Machine Learning Approaches

**LSTM (Long Short-Term Memory):**
- A type of recurrent neural network suitable for sequence prediction.
- **Simple Example:**
  

```python
import numpy as np


data = data['Close'].values.reshape(-1, 1)  
time_steps = 1

# Preprocess data into X_train and y_train
def create_dataset(data, time_steps):
    X, y = [], []
    for i in range(len(data) - time_steps):
        X.append(data[i:(i + time_steps), 0])  # Use 0 for the column index if 'Close' is the first column
        y.append(data[i + time_steps, 0])
    return np.array(X), np.array(y)

# Create sequences for LSTM
X_train, y_train = create_dataset(data, time_steps)

# Reshape X_train for LSTM input: [samples, time steps, features]
X_train = np.reshape(X_train, (X_train.shape[0], 1, X_train.shape[1]))

# Print shapes to verify
print('X_train shape:', X_train.shape)
print('y_train shape:', y_train.shape)
```

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense

model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(1, 1)))
model.add(LSTM(50))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(X_train, y_train, epochs=100, batch_size=32)
```

**Explanation and Example:**
- Identify unusual patterns that do not conform to expected behavior.
- **Example:**

```python
from sklearn.ensemble import IsolationForest

clf = IsolationForest(contamination=0.01)
data['Anomaly'] = clf.fit_predict(data[['Close']])
anomalies = data[data['Anomaly'] == -1]
plt.plot(data['Close'])
plt.scatter(anomalies.index, anomalies['Close'], color='red')
plt.title('Anomaly Detection')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```
